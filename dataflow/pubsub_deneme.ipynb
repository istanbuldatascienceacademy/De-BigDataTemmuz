```python
# Avro Schema
{
    "type": "record",
    "name": "Employee",
    "fields": [
        {"name": "Education", "type": "string"},
        {"name": "JoiningYear", "type": "int"},
        {"name": "City", "type": "string"},
        {"name": "PaymentTier", "type": "int"},
        {"name": "Age", "type": "int"},
        {"name": "Gender", "type": "string"},
        {"name": "EverBenched", "type": "string"},
        {"name": "ExperienceInCurrentDomain", "type": "int"},
        {"name": "LeaveOrNot", "type": "int"}
    ]        
}

import time 
from google.cloud import pubsub_v1
import pandas as pd
import os

os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "scenic.json"
project_id = "scenic-parity-429506-b8"
topic_name = "de_nisan2"

publisher = pubsub_v1.PublisherClient()
topic_path = publisher.topic_path(project_id, topic_name)

df = pd.read_csv("Employee.csv")
for index, row in df.iterrows():
    mesaj = row.to_json()
    gonder = publisher.publish(topic=topic_path, data=mesaj.encode("utf-8"))
    print(mesaj)
    time.sleep(1)
```
