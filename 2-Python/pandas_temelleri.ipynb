## DataFrame
Pandas DataFrame, verileri düzenlemek ve analiz etmek için kullanılan bir veri yapısıdır. 

DataFrame, verileri bir tablo şeklinde düzenler ve bu tablo üzerinde işlemler yapmamızı sağlar. 

DataFrame, satır ve sütunlardan oluşur. Satırlar, veri noktalarını temsil ederken, sütunlar veri noktalarının özelliklerini temsil eder(Kolonlar). 

DataFrame, farklı veri tiplerini destekler.


````python
veri = [["Yusuf", 50], ["Mehmet", 60], ["Ahmet", 70], ["Ali", 80], ["Veli", 90], ["Ayşe", 100], ["Fatma", 100]]
print(veri)
print(type(veri))
````

### Listeler den DataFrame Oluşturma
- Listelerden DataFrame oluşturmak için `pd.DataFrame()` fonksiyonunu kullanırız.

- DataFrame oluştururken, sütun isimlerini ve sütunlara ait verileri belirtmemiz gerekir.

- Sütun isimlerini belirtmek için `columns` parametresini kullanırız.



````python
df = pd.DataFrame(veri)
print(df)
df
````

**Not**: Eğer Dataframe de kolon isimleri verilmezse, kolon isimleri 0,1,2,3,... şeklinde verilir.

**Not**: Eğer Dataframe de kolon isimleri verilirse, kolon isimleri verilen isimlerle verilir.


````python
df = pd.DataFrame(veri, columns=["İsim", "Not"])
print(df)
df
````

````python
df = pd.DataFrame(veri, columns=["İsim", "Not"], index=["1", "2", "3", "4", "5", "6", "7"])
print(df)
````

### Dictionary den DataFrame Oluşturma
- Dictionary den DataFrame oluşturmak için `pd.DataFrame()` fonksiyonunu kullanırız.

- Dictionary oluştururken, keyler sütun isimlerini ve valuelar sütunlara ait verileri temsil eder.

- Sütun isimlerini belirtmek için `columns` parametresini kullanırız ama yine aynı isimleri vermek zorundayız.

- Sütun isimlerini belirtmezsek, sütun isimleri dictionary deki keyler olur.



````python
veri_tek = {
    "ad": "Yusuf",
    "not": 50
}

veri_cok = {
    "ad": ["Yusuf", "Mehmet", "Ahmet", "Ali", "Veli", "Ayşe", "Fatma"],
    "not": [50, 60, 70, 80, 90,15 ,100]
}

df1 = pd.DataFrame(veri_tek, index=[0])
print(df1)


df2 = pd.DataFrame(veri_cok)
print(df2)

df3 = pd.DataFrame(veri_cok, columns=["ad", "not"])
print(df3)
````

### Veriyi Tanıma ve Veriyi Düzenleme/Manipülasyon

### Filtreleme İşlemleri
- dataframe[condition] : dataframe'i belirtilen koşula göre filtreler.
- dataframe[condition1 & condition2] : dataframe'i birden fazla koşula göre filtreler.


````python
df = pd.read_csv("data/imdb_1000.csv")  
df.head(3)
````

````python
df.info()
````

````python
df["star_rating"]
````

````python
df["star_rating"] > 9.0
````

````python
df[df["star_rating"] > 9.0]
````

````python
df[
    (df["star_rating"] > 8.6) &
    (df["genre"] == "Crime") &
    (df["duration"] > df["duration"].mean() )
]
````

````python
df.describe()
````

````python
df[
    (df["star_rating"] > 9.0) & # and
    (df["duration"] > 120) | # or
    (df["duration"] < 70)
][["title", "genre"]]
````

````python
df[
    (df["star_rating"] > 9.0) & # and
    (df["duration"] > 120) | # or
    (df["duration"] < 70)
][["title", "genre"]].reset_index(drop=True)
````

### Gruplama İşlemleri
- ``dataframe.groupby()`` : dataframe'i belirtilen kolona göre gruplar.
- ``dataframe.groupby().mean()`` : dataframe'i belirtilen kolona göre gruplar ve ortalamasını alır.
- ``dataframe.groupby().sum()`` : dataframe'i belirtilen kolona göre gruplar ve toplamını alır.
- ``dataframe.groupby().count()`` : dataframe'i belirtilen kolona göre gruplar ve sayısını alır.
- ``dataframe.groupby().max()`` : dataframe'i belirtilen kolona göre gruplar ve maksimumunu alır.
- ``dataframe.groupby().min()`` : dataframe'i belirtilen kolona göre gruplar ve minimumunu alır.


````python
# Action türündeki filimlerin sayısı kaçtır?
df[df["genre"] =="Action"].count()
````

````python
# star_rating'i 8 den büyük olan ve Action türündeki filimlerin sayısı kaçtır?
df[
    (df["genre"] =="Action") &
    (df["star_rating"] > 8.0)
    ].count()
````

````python
df.groupby("genre").count()
````

````python
df.groupby("genre")[["title","star_rating"]].count()
````

````python
# star_rating'i 8 den büyük olan ve Action türündeki filimlerin genre ye count() uygulanmış hali
df[
    (df["genre"] =="Action") &
    (df["star_rating"] > 8.0)
    ].groupby("genre").count()
````

````python
# star_rating'i 8 den büyük olan ve Action türündeki filimlerin genre ye count() yap ve sadece title ve genre sütunlarını getir
df[
    (df["genre"] =="Action") &
    (df["star_rating"] > 8.0)
    ][["title", "genre"]].groupby("genre").count()


````
